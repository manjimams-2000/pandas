Calculate the median income by education level and gender.

```python
import pandas as pd

# Sample data for demonstration
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Education': ['Bachelor', 'Master', 'PhD', 'High School', 'Bachelor', 'Master', 'PhD', 'High School'],
    'Income': [50000, 60000, 70000, 40000, 55000, 65000, 75000, 45000]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Group by education level and gender, then calculate the median income
median_income = df.groupby(['Education', 'Gender'])['Income'].median().reset_index()

# Rename columns for better readability
median_income.columns = ['Education Level', 'Gender', 'Median Income']

print(median_income)
```
